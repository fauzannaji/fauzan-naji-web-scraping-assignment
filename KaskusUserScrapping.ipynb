```python
import requests
import json
from bs4 import BeautifulSoup
import pandas as pd
import requests
url = 'https://www.kaskus.co.id/@fufufafa/viewallposts/1/?sort=desc'
response = requests.get(url)
```

```python
response.content
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
posts = soup.find_all('div', class_='P(16px) Bgc(c-white)')

data = []

for post in posts:
    title_tag = post.find('div', class_='Fw(500) C(c-primary) Fz(18px) Mb(8px)')
    title = title_tag.get_text(strip=True)
    
    time_tag = post.find('span', class_='Fz(12px) C(c-secondary)')
    time = time_tag.get_text(strip=True) 
    
    content_tag = post.find('div', class_='C(c-secondary) Lh(1.5)')
    content = content_tag.get_text(strip=True) 
    
    data.append({'title': title, 'time': time, 'content': content})

```

```python
df = pd.DataFrame(data)
df
```

```python
df.to_csv('kaskus-user-post.csv' , index=False)
```
